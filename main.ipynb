```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datajkt1 = pd.read_csv('dataset\ispu_dki1.csv')
datajkt2 = pd.read_csv('dataset\ispu_dki2.csv')
datajkt3 = pd.read_csv('dataset\ispu_dki3.csv')
datajkt4 = pd.read_csv('dataset\ispu_dki4.csv')
datajkt5 = pd.read_csv('dataset\ispu_dki5.csv')
```

```python
display(datajkt1.head())
```

```python
# Set plot style
sns.set(style="whitegrid")

# Cleaning: Convert 'tanggal' to datetime format, and fill missing 'pm25' values with mean for visualizations
datajkt1['tanggal'] = pd.to_datetime(datajkt1['tanggal'])
datajkt1['pm25'] = datajkt1['pm25'].fillna(datajkt1['pm25'].mean())

# Creating a scatter plot of PM10 vs PM2.5
plt.figure(figsize=(10, 6))
sns.scatterplot(data=datajkt1, x='pm10', y='pm25')
plt.title('Scatter Plot of PM10 vs PM2.5 Concentrations')
plt.xlabel('PM10 Concentration')
plt.ylabel('PM2.5 Concentration')
plt.show()
```

Dari plot tersebut, terlihat sejumlah pengamatan titik data yang menunjukkan hubungan antara konsentrasi PM10 dan PM2.5 di udara.
- Sumbu X mewakili konsentrasi PM10 (Particulate Matter dengan diameter kurang dari atau sama dengan 10 mikrometer) dalam satuan yang tidak ditentukan.
- Sumbu Y mewakili konsentrasi PM2.5 (Particulate Matter dengan diameter kurang dari atau sama dengan 2.5 mikrometer) dalam satuan yang sama.

Dari scatter plot, kita dapat mengamati beberapa hal:
1. Ada sejumlah titik data yang mengindikasikan bahwa ketika konsentrasi PM10 meningkat, konsentrasi PM2.5 juga cenderung meningkat. Ini bisa menunjukkan adanya korelasi positif antara dua ukuran partikel tersebut.
2. Sebagian besar data terkonsentrasi di area konsentrasi PM10 yang lebih rendah, yang menunjukkan bahwa kondisi dengan konsentrasi PM10 yang lebih tinggi tidak sering terjadi atau tidak sering diukur dalam dataset ini.
3. Terdapat sejumlah titik yang menyebar yang dapat menunjukkan adanya variasi atau pengaruh lain dalam konsentrasi PM2.5 yang tidak sepenuhnya dijelaskan oleh konsentrasi PM10 saja.

```python
# Creating a histogram of CO concentrations
plt.figure(figsize=(10, 6))
sns.histplot(datajkt1['co'], kde=True, color='blue')
plt.title('Histogram of CO Concentrations')
plt.xlabel('CO Concentration')
plt.ylabel('Frequency')
plt.show()
```

Berikut merupakan visualisasi histogram dari konsentrasi karbon monoksida (CO) di udara dengan kurva yang menunjukkan distribusi teoritis dari data yang telah diolah.

- Sumbu X mewakili konsentrasi CO, biasanya diukur dalam satuan seperti bagian per juta (ppm) atau miligram per meter kubik (mg/m³), meskipun satuan sebenarnya tidak ditentukan di sini.
- Sumbu Y mewakili frekuensi, atau jumlah pengamatan, pada setiap interval konsentrasi CO.

Dari histogram, beberapa karakteristik distribusi data bisa diinterpretasikan:
1. Bentuk histogram menunjukkan sebagian besar pengamatan berkumpul di sekitar nilai konsentrasi tertentu yang lebih rendah, yang kemungkinan besar adalah mode dari distribusi.
2. Distribusi data tampaknya asimetris dengan "ekor" yang memanjang ke arah nilai konsentrasi yang lebih tinggi, menunjukkan bahwa ada sejumlah kecil pengamatan dengan konsentrasi CO yang sangat tinggi.
3. Kurva yang ditumpangkan pada histogram mungkin merupakan kurva distribusi normal atau distribusi probabilitas lain yang digunakan untuk menggambarkan bentuk teoritis dari data yang teramati.

```python
# Boxplot for pollutant distribution by air quality category
plt.figure(figsize=(12, 8))
sns.boxplot(x='categori', y='pm10', data=datajkt1)
plt.title('Boxplot of PM10 Distribution by Air Quality Category')
plt.xlabel('Air Quality Category')
plt.ylabel('PM10 Concentration')
plt.show()
```


Berikut merupakan visualisasi boxplot dari distribusi konsentrasi PM10 berdasarkan kategori kualitas udara.

- Sumbu X mengkategorikan kualitas udara menjadi empat kategori: 'SEDANG', 'BAIK', 'TIDAK SEHAT', dan 'TIDAK ADA DATA'.
- Sumbu Y menunjukkan konsentrasi PM10, biasanya diukur dalam mikrogram per meter kubik (µg/m³), meskipun satuan sebenarnya tidak ditentukan.

Dari boxplot, kita dapat mengamati beberapa hal:
1. Kategori 'SEDANG': Median konsentrasi PM10 lebih tinggi dibandingkan dengan kategori 'BAIK', dengan rentang interkuartil yang lebih luas, yang menunjukkan variabilitas yang lebih besar dalam konsentrasi PM10 pada hari-hari dengan kualitas udara 'SEDANG'.
2. Kategori 'BAIK': Boxplot ini lebih pendek, yang menunjukkan variabilitas konsentrasi PM10 yang lebih rendah pada hari-hari dengan kualitas udara yang baik. Median konsentrasi PM10 di kategori ini lebih rendah daripada di kategori 'SEDANG'.
3. Kategori 'TIDAK SEHAT': Memiliki median yang lebih tinggi dibandingkan dengan dua kategori sebelumnya, yang mengindikasikan bahwa konsentrasi PM10 cenderung lebih tinggi pada hari-hari yang kualitas udaranya dikategorikan sebagai 'TIDAK SEHAT'.
4. Kategori 'TIDAK ADA DATA': Tidak menunjukkan data yang cukup untuk membuat kesimpulan, atau bisa juga berarti bahwa tidak ada pengukuran yang tersedia atau dilakukan untuk periode atau kategori ini.

Titik-titik yang berada di luar 'whiskers' pada boxplot menandakan pencilan (outliers), yaitu nilai yang secara signifikan lebih tinggi atau lebih rendah daripada data lainnya.

```python
# Time series plot for PM10 and PM2.5
plt.figure(figsize=(12, 8))
sns.lineplot(data=datajkt1, x='tanggal', y='pm10', label='PM10', color='red')
sns.lineplot(data=datajkt1, x='tanggal', y='pm25', label='PM2.5', color='blue')
plt.title('Time Series of PM10 and PM2.5 Concentrations')
plt.xlabel('Date')
plt.ylabel('Concentration')
plt.legend()
plt.show()
```

Berikut merupakan visualisasi time series atau deret waktu dari konsentrasi PM10 dan PM2.5 selama beberapa tahun.

- Sumbu X mewakili waktu, dengan label tahun yang terbaca dari 2010 hingga 2022, menunjukkan periode pengamatan.
- Sumbu Y menunjukkan konsentrasi PM10 dan PM2.5, yang biasanya diukur dalam mikrogram per meter kubik (µg/m³), meskipun satuan tidak ditentukan secara spesifik.

Dari time series, kita bisa mengamati beberapa hal:

1. Konsentrasi PM10 (garis merah): Terdapat fluktuasi konsentrasi PM10 selama periode yang diamati, dengan beberapa puncak yang menunjukkan peningkatan konsentrasi. Garis merah yang terus-menerus menandakan data PM10 yang dicatat sepanjang waktu.
2. Konsentrasi PM2.5 (garis biru): Seperti PM10, konsentrasi PM2.5 juga menunjukkan variasi seiring waktu dengan puncak dan lembah yang mungkin mencerminkan perubahan musiman atau peristiwa tertentu yang mempengaruhi kualitas udara.
3. Garis horisontal: Garis ini mungkin mewakili nilai rata-rata atau median dari masing-masing konsentrasi partikel selama periode yang ditunjukkan, memudahkan identifikasi kapan konsentrasi partikel berada di atas atau di bawah rata-rata.

```python
# Bar plot of number of critical pollution days by pollutant
critical_counts = datajkt1['critical'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=critical_counts.index, y=critical_counts.values, hue=critical_counts.index, palette='viridis', legend=False)
plt.title('Number of Critical Pollution Days by Pollutant')
plt.xlabel('Pollutant')
plt.ylabel('Count of Critical Days')
plt.show()
```

Berikut merupakan visualisasi barplot yang menggambarkan jumlah hari dengan polusi kritis berdasarkan jenis polutan.
- Sumbu X mewakili jenis polutan, yang terdiri dari PM10, O3 (ozon), PM2.5, CO (karbon monoksida), dan SO2 (dioksida belerang).
- Sumbu Y menunjukkan jumlah hari dimana tingkat polusi dianggap kritis, dengan angka yang tidak ditentukan secara jelas tetapi dapat diukur dalam jumlah hari.

Dari batang grafik, kita dapat mengamati beberapa hal:
1. PM10: Menunjukkan jumlah hari kritis yang paling tinggi dibandingkan dengan polutan lainnya, yang digambarkan dengan batang berwarna ungu.
2. O3: Ozon memiliki jumlah hari kritis yang hampir sama dengan PM10, tetapi sedikit lebih rendah, yang digambarkan dengan batang berwarna biru.
3. PM2.5: Memiliki jumlah hari kritis yang lebih sedikit dibandingkan dengan PM10 dan O3, yang digambarkan dengan batang berwarna hijau.
4. CO: Batang berwarna hijau muda untuk CO menunjukkan jumlah hari kritis yang jauh lebih sedikit dibandingkan dengan partikel lain.
5. SO2: Memiliki jumlah hari kritis yang terkecil dalam data yang digambarkan, yang ditunjukkan dengan batang berwarna hijau muda yang paling pendek.

```python
# Correlation matrix of pollutants
corr_matrix = datajkt1[['pm10', 'pm25', 'so2', 'co', 'o3', 'no2']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Pollutant Levels')
plt.show()
```

Berikut merupakan visualisasi matriks korelasi untuk berbagai level polutan di atmosfer.
- Sumbu X dan Y sama-sama menunjukkan jenis polutan: PM10, PM2.5, SO2 (dioksida belerang), CO (karbon monoksida), O3 (ozon), dan NO2 (dioksida nitrogen).
- Kotak-kotak dalam matriks menunjukkan koefisien korelasi antara pasangan polutan, dengan skala warna yang disediakan di sisi kanan, mulai dari -1 (korelasi negatif penuh, ditandai dengan biru tua) hingga +1 (korelasi positif penuh, ditandai dengan merah).

Dari matriks korelasi, kita dapat menarik beberapa kesimpulan:
1. Setiap polutan memiliki korelasi sempurna dengan dirinya sendiri, yang ditandai dengan nilai 1.00 dan warna merah tua, yang merupakan fitur standar dari matriks korelasi.
2. PM10 dan CO memiliki korelasi positif yang relatif tinggi (0.34), yang bisa menunjukkan bahwa keduanya cenderung meningkat bersamaan.
3. CO dan O3 memiliki korelasi negatif sedikit (-0.11), yang mungkin menunjukkan bahwa ketika konsentrasi CO meningkat, konsentrasi O3 cenderung sedikit menurun, atau sebaliknya.
4. NO2 memiliki korelasi positif yang moderat dengan PM10 (0.33) dan PM2.5 (0.22), yang mungkin menandakan hubungan antara sumber emisi dari kedua polutan tersebut.

Nilai korelasi yang mendekati 0 menunjukkan tidak adanya korelasi yang kuat antara pasangan polutan tersebut.
